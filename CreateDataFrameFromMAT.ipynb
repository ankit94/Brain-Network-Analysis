```python
import numpy as np
import pandas as pd
import scipy.io as sio
import os
import pickle
```

```python
path = r'C:\Users\ankit\Documents\Fall 19\CSE 575 SML\CSE575Brains-master\CSE575Brains-master\brainnetworks\smallgraphs'
path
```

```python
def upper_tri_indexing(A):
    m = A.shape[0]
    r,c = np.triu_indices(m,1)
    return A[r,c]

data_list = []
for r, d, f in os.walk(path):
    for file in f:
        mat_file = sio.loadmat(path + '\\' + file)
        data = mat_file['fibergraph'].todense()
        data = upper_tri_indexing(data)
        data = data.flatten()
        data = data.tolist()
        data[0].append(file[:-10])
        data_list.append(data[0])
```
